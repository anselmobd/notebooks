```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
# dataset
n = 1_000_000
df = pd.DataFrame({
    "valor": np.random.randint(0, 200, size=n)
})
```

```python
# 1. iterrows
df1 = df.copy()
t0 = time.time()

for i, row in df1.iterrows():
    if row["valor"] > 100:
        df1.loc[i, "novo"] = row["valor"] * 2
    else:
        df1.loc[i, "novo"] = row["valor"]

t_iterrows = time.time() - t0
```

```python
# 2. apply
df2 = df.copy()
t0 = time.time()

df2["novo"] = df2.apply(
    lambda row: row["valor"] * 2 if row["valor"] > 100 else row["valor"],
    axis=1
)

t_apply = time.time() - t0
```

```python
# 3. vetorizado
df3 = df.copy()
t0 = time.time()

df3["novo"] = np.where(
    df3["valor"] > 100,
    df3["valor"] * 2,
    df3["valor"]
)

t_vectorized = time.time() - t0
```

```python
labels = ["iterrows", "apply", "vetorizado"]
tempos = [t_iterrows, t_apply, t_vectorized]

plt.figure(figsize=(8, 4))
plt.bar(labels, tempos)
plt.ylabel("Tempo (segundos)")
plt.title("Comparação de performance - Pandas")

for i, v in enumerate(tempos):
    plt.text(i, v, f"{v:.3f}s", ha="center", va="bottom")

plt.show()
```
